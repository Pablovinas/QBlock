```python
n_nodes = 4
nodes = list(range(1,n_nodes+1))

block_num = 2   # number of new block  

path_old = 'Blocks/block' + str(block_num-1) + '.csv'
old_block = pd.read_csv(path_old)

old_block.head()
```

```python
path_tran = 'Transactions/transactions.csv'
tran = pd.read_csv(path_tran)
tran.set_index('Node', inplace=True)


tran_1 = 'Transaction ' + str(2*(block_num-1)+1)
tran_2 = 'Transaction ' + str(2*block_num)


t_1 = [0,1,0,0,-1]
t_2 = [0,0,0,5,-5]

old_w = np.asarray(old_block['Wallet'].tolist())

# cambiar atributo de current block de la clase bloque? Definir el bloque con las transacciones iniciales
# y unos números iniciales
# funcion transaccion dentro de la clase bloque

node = 1

payment = 0.2*np.ones(5) 
payment[node-1]=1

new_w = list(old_w + np.asarray(t_1) + np.asarray(t_2) + payment)

new_block = pd.DataFrame({'Node': [1,2,3,4,5], 'Old Wallet': old_block['Wallet'], tran_1: list(t_1),
                          tran_2: list(t_2), 'New Wallet': new_w,
                   'Random number': ['01011','10110','11100','00000','10111'],
                   'prev_fidelity': [0,0.3,0.4,0.5,0.8]}, dtype=object)

new_block.set_index('Node',inplace=True)
new_block.to_csv('Blocks/block'+str(block_num)+'.csv')
new_block.head()

```

```python
# Añadimos al csv de transacciones las nuevas transacciones
path_transaction = 'Transactions/transactions.csv'
tran = pd.read_csv(path_transaction)
tran.set_index('Node', inplace=True)
tran[tran_1]=t_1
tran[tran_2]=t_2
tran.to_csv('Transactions/transactions.csv')
tran.head()
```
